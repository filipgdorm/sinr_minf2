```python
import pandas as pd
```

```python
df = pd.read_csv('./data/train/geo_prior_train.csv')
```

```python
grouped_df = df.groupby('taxon_id').size().reset_index(name='num_entries')
```

```python
grouped_df
```

```python
grouped_df.sort_values(by='num_entries', ascending=False)
```

```python
percentile_95 = grouped_df['num_entries'].quantile(0.95)
print(percentile_95)
```

```python
num_taxa_gt_10000 = grouped_df[grouped_df['num_entries'] < 500]
print("num entries with more than 10,000 obs: ", len(num_taxa_gt_10000))
```

```python
# see overlap with iucn data
iucn_taxa_df = pd.read_csv('./my_experiments/f1_score_baseline/results/theoretical_optimal_iucn_f1score.csv', index_col=0)
iucn_taxa_ids = set(iucn_taxa_df.taxon_id.values)
```

```python
inat_ids = set(num_taxa_gt_10000.taxon_id)
```

```python
# Find the overlap (common elements) between the two sets
overlap = inat_ids.intersection(iucn_taxa_ids)

# Find the elements unique to each set
unique_to_inat = inat_ids.difference(iucn_taxa_ids)
unique_to_iucn = iucn_taxa_ids.difference(inat_ids)

print("Overlap between inat_ids and iucn_taxa_ids size:", len(overlap), " overlap: ", overlap )
print("Elements unique to inat_ids: size:", len(unique_to_inat), " overlap: ", unique_to_inat )
print("Elements unique to iucn_taxa_ids: size:", len(unique_to_iucn), " overlap: ", unique_to_iucn )
```
