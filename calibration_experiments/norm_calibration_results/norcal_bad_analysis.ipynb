```python
import pandas as pd
```

```python
freq_df = pd.read_csv('../../data/train/geo_prior_train.csv')

counts = freq_df.taxon_id.value_counts()
```

```python
bins = [0, 100, 1000, float('inf')]
labels = ['rare', 'common', 'frequent']
```

```python
df_anfull_norcal = pd.read_csv('./an_full_-1/0.1/scores.csv')
```

```python
df_anfull = pd.read_csv('../../new_results/an_full_-1/scores.csv', index_col="taxon_id").drop(columns=['Unnamed: 0'])
df_anfull['count'] = counts
df_anfull = df_anfull.reset_index()
df_anfull['tace'] = pd.read_csv('../calibration_metrics_results/an_full_-1/scores.csv')[['tace']]
df_anfull['tace_calib'] = pd.read_csv('./an_full_-1/-0.1/scores.csv')['tace']
df_anfull['opt_thres'] = pd.read_csv('./an_full_-1/-0.1/scores.csv')['opt_thres']
df_anfull['fscore50'] = pd.read_csv('./an_full_-1/0.1/scores_v2.csv')['fscore50']


df_anfull['count_category'] = pd.cut(df_anfull['count'], bins=bins, labels=labels, right=False)
grouped_df = df_anfull.groupby('count_category').agg({'thres': 'mean', 'fscore': 'mean', 'tace': 'mean', 'tace_calib': 'mean','opt_thres': 'mean', 'fscore50': 'mean', 'taxon_id': 'count'})
grouped_df["tace_gap"] = grouped_df['tace_calib'] - grouped_df['tace']
grouped_df
```
